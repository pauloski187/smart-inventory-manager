# Inventory Exploratory Data Analysis

This notebook focuses on inventory-related analysis from the orders data.

```python
# Load and analyze inventory data
import pandas as pd
import matplotlib.pyplot as plt

# Load processed data
df = pd.read_csv('../data/processed/amazon_orders_processed.csv')

# Calculate inventory metrics
product_metrics = df.groupby('ProductID').agg({
    'Quantity': 'sum',
    'TotalAmount': 'sum',
    'OrderID': 'count'
}).rename(columns={'OrderID': 'order_count'})

print("Top 10 products by sales volume:")
print(product_metrics.sort_values('Quantity', ascending=False).head(10))
```
